```python
import csv
from matplotlib import pyplot as plt
import numpy as np
from os import listdir
from os.path import isfile, join
import math
import matplotlib.image as mpimg

import os 
dir_path = os.path.dirname(os.path.realpath('__file__'))
onlyfiles = [f for f in listdir('LIDAR Data') if isfile(join('LIDAR Data', f))]
```

```python
onlyfiles
onlyfiles = np.sort(onlyfiles)
mean1 = []
angles = []
mean2 = []
i = 0
for file in onlyfiles:
    file_name = join(dir_path,'LIDAR Data',file)
    
    with open(file_name, 'r') as f:
        reader = csv.reader(f)
        your_list = list(reader)
        integers = []
        for i in range(len(your_list)):
            string = ((your_list[i][0]))
            integers.append(int(string))
        r_temp = np.mean(integers)
        if 'Lidar_1' in file_name:
            mean1.append(r_temp)
        elif 'Lidar_2' in file_name:
            mean2.append(r_temp)
        i+=1
        
for i in range(len(mean1)):
    angles.append(math.radians(i * 45))
```

```python
def plot_fig(angles, means):
    plt.figure(figsize=(20,10))
    ax = plt.subplot(111, projection='polar') # plotting the room data
    ax.plot(angles,means,".",markersize=10)
    ax.set_title("Polar Graph", va='bottom')
    plt.show()
```

```python
plot_fig(angles,mean1)
plot_fig(angles,mean2)
```
