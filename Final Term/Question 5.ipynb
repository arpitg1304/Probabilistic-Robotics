```python
# -*- coding: utf-8 -*-
"""
Created on Wed Dec 13 16:59:10 2017

@author: Arpit
"""

# -*- coding: utf-8 -*-
"""
"""

import numpy as np
import csv

#Conversion of gps code to degrees
def code_to_deg(gps_data):

    wholeDeg = np.floor(gps_data/100.0)
    minutes = gps_data - 100*wholeDeg
    total = wholeDeg + (minutes/60.0)
    return total

def read_data(filename):
    
    with open (filename,'r') as csvfile:
        plots = csv.reader(csvfile)
        read_list = list(plots)
    time = []
    lat = []
    long = []
    temp = 1
    for i in read_list:
            if temp>1:
                 time.append(float(i[0]))
                 lat.append(float(i[1]))
                 long.append(float(i[3]))
            temp = temp+1
    time = np.array(time)
    lat = np.array(lat)
    long = np.array(long)

    position = np.array([lat,long])
    position = position.T
    position = code_to_deg(position)
    return time,position

print('Answers to Part A')
print('\n')

Master_time,Master_pos = read_data("Master.csv")
Average_Ma_pos = np.mean(Master_pos,axis=0)
print('Average location of Master: Latitude ', Average_Ma_pos[0], 'Longitude', Average_Ma_pos[1])
print('\n')

Rover1_time,Rover1_pos = read_data("Rover_1.csv")
m_r1_pos = np.mean(Rover1_pos,axis = 0)
print('Average location of Rover1: Latitude ', m_r1_pos[0], 'Longitude', m_r1_pos[1])
print('\n')

Rover2_time,Rover2_pos = read_data("Rover_2.csv")
m_r2_pos = np.mean(Rover2_pos,axis = 0)
print('Average location of Rover2: Latitude ', m_r2_pos[0], 'Longitude', m_r2_pos[1])
print('\n')

print('Answers to Part B')
print('\n')

Masterlatitude = 111073.9415784558
Masterlongitude = 82804.80193609379
Mconversion =[Masterlatitude, Masterlongitude]
Mconv=np.array(Mconversion)

M_error = (Master_pos-Average_Ma_pos)*Mconv
M_error_std = np.std(M_error,axis = 0)

print('Standard Deviation of error for Master(in metres): Latitude-',M_error_std[0],'Longitude-',M_error_std[1])
print('\n')


R1latitude = 111073.94016980016
R1longitude = 82804.89609438839
R1conversion = [R1latitude, R1longitude]
R1conv=np.array(R1conversion)
 
R1_error = (Rover1_pos-m_r1_pos)*R1conv
R1_error_std = np.std(R1_error,axis = 0)

print('Standard Deviation of error for Rover1(in metres): Latitude-',R1_error_std[0],'Longitude-',R1_error_std[1])
print('\n')
R2latitude = 111073.93846804278
R2longitude = 82805.00984424443
R2conversion = [R2latitude, R2longitude]
R2conv=np.array(R2conversion)

R2_error = (Rover2_pos-m_r2_pos)*R2conv
R2_error_std = np.std(R2_error,axis = 0)

print('Standard Deviation of error for Rover2(in metres): Latitude-',R2_error_std[0],'Longitude-',R2_error_std[1])
print('\n')

R1_Mconv = (R1conv+Mconv)/2
R2_Mconv = (R2conv+Mconv)/2
                
    
R1_M_Distance = np.sqrt(np.sum(np.square((m_r1_pos-Average_Ma_pos)*R1_Mconv)))
R2_M_Distance = np.sqrt(np.sum(np.square((m_r2_pos-Average_Ma_pos)*R2_Mconv)))

print('Answers to Part C')
print('\n')
    
print('Distance of Rover1 to Master in metres:',R1_M_Distance)
print('\n')
print('Distance of Rover2 to Master in metres:',R2_M_Distance)
print('\n')


```
